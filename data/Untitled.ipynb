```python
import pandas as pd
import numpy as np

grades = [48, 99, 75, 80, 42, 80, 72, 68, 36, 78]

df = pd.DataFrame(

    {

        "ID": ["x%d" % r for r in range(10)],

        "Gender": ["F", "M", "F", "M", "F", "M", "F", "M", "M", "M"],

        "ExamYear": [

            "2007",

            "2007",

            "2007",

            "2008",

            "2008",

            "2008",

            "2008",

            "2009",

            "2009",

            "2009",

        ],

        "Class": [

            "algebra",

            "stats",

            "bio",

            "algebra",

            "algebra",

            "stats",

            "stats",

            "algebra",

            "bio",

            "bio",

        ],

        "Participated": [

            "yes",

            "yes",

            "yes",

            "yes",

            "no",

            "yes",

            "yes",

            "yes",

            "yes",

            "yes",

        ],

        "Passed": ["yes" if x > 50 else "no" for x in grades],

        "Employed": [

            True,

            True,

            True,

            False,

            False,

            False,

            False,

            True,

            True,

            False,

        ],

        "Grade": grades,

    }

)
```

```python
df
```

```python
df.groupby("ExamYear").agg(
    {
        "Participated": lambda x: x.value_counts()["yes"],
        "Passed": lambda x: sum(x == "yes"),
        "Employed": lambda x: sum(x),
        "Grade": lambda x: sum(x) / len(x),
    }
)
```
